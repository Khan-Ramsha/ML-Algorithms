# Time resampling

```python
df.resample(rule="A").min()["Open"].plot() 
```

```python
## rule A - year end frequency
df.resample(rule="A").max()["Open"].plot()
```

```python
df["Open"].resample(rule="A").mean().plot(kind="bar")
```

```python
df["Open"].resample(rule="M").max().plot(kind='bar',figsize=[15,6])
```

## Simple Moving Average

```python
df["Open:10DaysRolling"]=df["Open"].rolling(window=10,min_periods=2).mean()
```

```python
df["Open:10DaysRolling"].head()
```

```python
df[["Open","Open:10DaysRolling"]].plot(figsize=(15,4))
```

```python
# Cummulative moving avg
df["Open"].expanding().mean().plot(figsize=(15,4))
```

```python
df["EMA:Open"]=df["Open"].ewm(alpha=0.1,adjust=False).mean()
```

```python
df["EMA:Open"]
```

```python
# exponential moving avg-smoothing purpose and giving more weight(importance) to recent observation
df[["Open","EMA:Open"]].plot(figsize=(15,4))
```

```python
# exponential weighted moving avg- we use ewm() only but using adjust=true
df["EWMA:Open"]=df["Open"].ewm(alpha=0.1,adjust=True).mean() # adjust=true is default
```

```python
df[["Open","EWMA:Open"]].plot(figsize=(15,4))
```
